```python
import numpy as np
import matplotlib.pyplot as plt
import sounddevice as sd

A = 1 #amplitude
fsin = 1000 #fréquence de la sinusoïde
fe = 100000 #fréquence d'échantillonnage

#durée
D = 5
t = np.arange(0, D, 1/fe) #temps

signal = A * np.cos(fsin * 2 * np.pi * t) #sinusoïde 

#tracé de la sinusoïde créée
fig = plt.figure(figsize=(12, 6))
ax = plt.axes()
ax.plot(t,signal)
plt.xlim(0, 0.003)

fig = plt.figure(figsize=(12, 6))

plt.stem(signal[0:50])

sd.play(signal, fe, loop=True)
```

```python
N = fe/fsin

def erreur(array, N):
    max = np.sin(np.pi/2)
    erreur = 0
    ecart = 0
    for i in range(len(array)):
        if array[i] == max:
            print("ok")
        else:
            a = np.sin(np.pi/2 + np.pi/N)
            erreur = abs((a - 1) * 100)
            return erreur
err = erreur(t, N)
print("l'erreur est de ", err,"%")
```
